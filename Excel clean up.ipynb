# Final

```python
import os
import pandas as pd
```

```python
original_directory = "Originals/"
original_files = os.listdir(original_directory)
```

```python
directory_2019 = "2019/"
directory_2020 = "2020/"
```

```python
for i in original_files:
    data = pd.read_excel( os.path.join(original_directory, i) )
    data["Release Date"] = pd.to_datetime(data["Release Date"])
    
    data2020 = data[(data['Release Date'] >= '2020-01-1') & (data['Release Date'] <= '2020-12-31')]
    data2019 = data[(data['Release Date'] >= '2019-01-1') & (data['Release Date'] <= '2019-12-31')]
    
    data2020.to_excel(os.path.join(directory_2020, i.split()[0] + "_ene-sep2020.xlsx"), index=False)
    data2019.to_excel(os.path.join(directory_2019, i.split()[0] + "_ene-dic2019.xlsx"), index=False)
```

```python
numero_de_casos = amikacina2019["Number of Cases"].fillna(1).sum()
numero_de_reportes = amikacina2019.shape[0]
numero_de_offlabel = amikacina2019[amikacina2019["Record Type"] == "Special Situation Case Report"].shape[0]
pais_con_mas_reportes = amikacina2019["Reporter Country"].value_counts().index[0]
top_age_group = amikacina2019["Age Group"].value_counts().index[0]
numero_de_serios = amikacina2019[amikacina2019["Serious"] == "Yes"].shape[0]
numero_de_no_serios = amikacina2019[amikacina2019["Serious"] == "No"].shape[0]
numero_de_first_reports = amikacina2019[amikacina2019["First Report"] == "Yes"].shape[0]
numero_de_no_first_reports = amikacina2019[amikacina2019["First Report"] == "No"].shape[0]
```

```python
nombres = ["Casos", "Reportes", "Off Label", "Pais con mas Reportes", 
           "Top Age Group", "Casos Serios", "Casos No Serios", "First Reports", 
           "No First Reports"]

valores = [numero_de_casos, numero_de_reportes, numero_de_offlabel, pais_con_mas_reportes, top_age_group,
           numero_de_serios, numero_de_no_serios, numero_de_first_reports, numero_de_no_first_reports]
```

```python
final_table = pd.DataFrame({
    "Descripcion": nombres,
    "Valores": valores
})
```

```python
def get_data(data):
    numero_de_casos = data["Number of Cases"].fillna(1).sum()
    numero_de_reportes = data.shape[0]
    numero_de_offlabel = data[data["Record Type"] == "Special Situation Case Report"].shape[0]
    pais_con_mas_reportes = data["Reporter Country"].value_counts().index[0]
    top_age_group = data["Age Group"].fillna("Desconocido").value_counts().index[0]
    numero_de_serios = data[data["Serious"] == "Yes"].shape[0]
    numero_de_no_serios = data[data["Serious"] == "No"].shape[0]
    numero_de_first_reports = data[data["First Report"] == "Yes"].shape[0]
    numero_de_no_first_reports = data[data["First Report"] == "No"].shape[0]
    
    nombres = ["Casos", "Reportes", "Off Label", "Pais con mas Reportes", 
           "Top Age Group", "Casos Serios", "Casos No Serios", "First Reports", 
           "No First Reports"]

    valores = [numero_de_casos, numero_de_reportes, numero_de_offlabel, pais_con_mas_reportes, top_age_group,
               numero_de_serios, numero_de_no_serios, numero_de_first_reports, numero_de_no_first_reports]
    
    final_table = pd.DataFrame({
        "Descripcion": nombres,
        "Valores": valores
    })
    
    return final_table
```

```python
original_directory2019 = "2019/"
original_files2019 = os.listdir(original_directory2019)
```

```python
original_directory2020 = "2020/"
original_files2020 = os.listdir(original_directory2020)
```

```python
directory_2019 = "2019_resumen/"
directory_2020 = "2020_resumen/"
```

```python
for i in original_files2019:
    print(i)
    if i == ".DS_Store" or i.startswith("~"):
        continue
        
    data = pd.read_excel( os.path.join(original_directory2019, i) )
    
    if data.shape[0] == 0:
        continue
    
    final_data2019 = get_data(data)
    final_data2019.to_excel(os.path.join(directory_2019, i.split("_")[0] + "_resumen2019.xlsx"), index=False)
```

```python
for i in original_files2020:
    print(i)
    if i == ".DS_Store" or i.startswith("~"):
        continue
        
    data = pd.read_excel( os.path.join(original_directory2020, i) )
    
    if data.shape[0] == 0:
        continue
    
    final_data2020 = get_data(data)
    final_data2020.to_excel(os.path.join(directory_2020, i.split("_")[0] + "_resumen2020.xlsx"), index=False)
```

```python
import numpy as np
from shutil import copyfile
```

```python
original_directory2020 = "2020/"
original_files2020 = os.listdir(original_directory2020)

covid_2020_directory = "2020_covid/"
```

```python
keywords = ["coronavirus", "covid", "sars-cov2", "off label", "covid-19"]
```

```python
for i in original_files2020:
    if i == ".DS_Store" or i.startswith("~"):
        continue
        
    data = pd.read_excel( os.path.join(original_directory2020, i) )
    
    if data.shape[0] == 0:
        continue
    
    for word in keywords:
        data[word] = data.apply(lambda row: row.astype(str).str.contains(word.lower(), case=False).any(), axis=1)
        
        
    if data.iloc[:, -len(keywords):].sum().values.sum() > 0:
        print(i)
        copyfile(os.path.join(original_directory2020, i), os.path.join(covid_2020_directory, i))
```
